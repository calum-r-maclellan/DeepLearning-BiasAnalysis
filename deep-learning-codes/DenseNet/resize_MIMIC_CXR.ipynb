```python
# given a subset of the data, make a new metadata file corresponding to the patient subset
img_subset_path = preproc_dir
df_metadata_subset = pd.DataFrame(columns=df_metadata.columns)
i=0
files=[]
for file in os.listdir(img_subset_path):
    i+=1
    if i==1:
        f = file.strip('_resized.jpg')
        df_f = df_metadata.loc[df_metadata['dicom_id']==f, :]
        df_metadata_subset = pd.concat([df_metadata_subset, df_f], ignore_index=True)
        files.append(file)
        # display(df_metadata_subset)
    elif file in files:
        print('duplicate')
        continue
    else:
        f = file.strip('_resized.jpg')
        df_f = df_metadata.loc[df_metadata['dicom_id']==f, :]
        df_metadata_subset = pd.concat([df_metadata_subset, df_f], ignore_index=True)
        files.append(file)

    if (i%1000)==0:
        print(i)

```

```python
df_metadata_subset = df_metadata_subset.drop_duplicates('dicom_id')
display(df_metadata_subset)
```
